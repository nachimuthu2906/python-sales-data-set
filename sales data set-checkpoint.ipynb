EXPLORATORY DATA ANALYSIS

GENDER

```python
# plotting a bar chart for gender and its count
sns.countplot(x = 'Gender', y = None, data = df, hue = "Gender" )
plt.title("Total_Gender")
plt.show()
sns.countplot()
```

```python
# plotting a bar chart for gender vs total amount
grouped_df = df.groupby('Gender', as_index=False)['Amount'].sum()
sns.barplot(x = 'Gender', y = 'Amount',  data = grouped_df, hue= "Gender")
plt.title("Gender vs Total_Amount")
plt.show()
```

```python
# plotting a bar chart for agegroub vs Gender
grouped_df = df.groupby(['Age Group', 'Gender']).size().reset_index(name='Count')
sns.barplot(x='Age Group', y='Count', hue='Gender', data=grouped_df)
plt.title("Age Group vs Gender Relationship")
plt.ylabel("Count")
plt.xlabel("Age Group")
plt.legend(title='Gender')
plt.show()
```

```python
# Plotting bar chart for Total Amount vs Age Group
grouped_df = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by = 'Amount')
sns.barplot(x='Age Group', y='Amount', data=grouped_df, hue = 'Age Group')
plt.title("Total Amount vs Age Group")
plt.ylabel("Total Amount")
plt.xlabel("Age Group")
plt.show()
```

```python
# Plotting bar chart for total number of orders from top 10 states
order_counts = df.groupby('State').size().reset_index(name='Total Orders')
top_states = order_counts.sort_values(by='Total Orders', ascending=False).head(10)
sns.barplot(x='State', y='Total Orders', data=top_states, hue='State', palette='viridis')
plt.title("Total Number of Orders from Top 10 States")
plt.ylabel("Total Orders")
plt.xlabel("State")
plt.xticks(rotation=90) 
plt.show()
```

```python
# Plotting scatter chart for total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.scatterplot(data=sales_state, x='State', y='Amount', s=200, color='blue', marker='o')
plt.title("Total Sales from Top 10 States", size = 40)
plt.ylabel("Total Sales Amount", size = 30)
plt.xlabel("State", size = 30)
plt.xticks(rotation= 90,size =20)
plt.show()
```

```python
# Plotting pie chart for Total marital status
plt.figure(figsize=(10,4))
df["Marital_Status"].value_counts().plot(kind="pie", autopct = "%1.2f%%")
plt.show()
```

```python
# Plotting bar chart for Total Sales by Marital Status and Gender
Sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum()
sns.set(rc={'figure.figsize': (10, 6)})
sns.barplot(data = Sales_state, x='Marital_Status', y='Amount', hue='Gender')
plt.title('Total Sales by Marital Status and Gender')
plt.ylabel('Total Sales Amount')
plt.xlabel('Marital Status')
plt.title('Gender')
plt.show()
```

```python
# plotting pie chart for count a total occupation
occupation_counts = df['Occupation'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(occupation_counts, labels=occupation_counts.index, autopct='%1.1f%%', startangle=60)
plt.title('Distribution of Occupations', size=20)
plt.axis('equal')  
plt.show()
```

```python
# plotting bar chart for total sales amount by Occupation
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation', y= 'Amount', hue= 'Occupation')
plt.title('Total Sales Amount by Occupation', size =40)
plt.xlabel('Occupation', size =30)
plt.ylabel('Total Sales Amount',size =30)
plt.xticks(rotation=90,size =20) 
plt.yticks(size = 20)
plt.show()
```

```python
# plotting bar chart for product category
sns.set(rc={'figure.figsize':(20,5)})
sns.countplot(data= df, x='Product_Category', hue= "Product_Category")
plt.title('product category', size =30)
plt.xlabel('Products', size =20)
plt.ylabel('Count',size =20)
plt.xticks(rotation = 90, size = 20)
plt.yticks(size = 20)
plt.show()
```

```python
# plotting bar chart for top 10 product category by sales
sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales, x='Product_Category', y='Amount',hue = 'Product_Category')
plt.title('Top 10 Product Categories by Sales Amount', size=30)
plt.xlabel('Product Categories', size=20)
plt.ylabel('Total Sales Amount', size=20)
plt.xticks(rotation = 90, size=20)  
plt.yticks(size=20)
plt.show()
```

```python
# plotting bar chart for top 10 sold products
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders', hue = 'Product_ID', palette='viridis')
plt.title('Top 10 Sold Products ', size = 40)
plt.xlabel('Product ID',size = 30)
plt.ylabel('Total Orders',size = 30)
plt.xticks(rotation=90, size = 20)
plt.yticks(size = 20)
plt.show()
```
